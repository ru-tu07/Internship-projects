I checked the dataset for missing values to make sure the data is complete before starting the analysis. After checking, I found that there are no missing values in the dataset, so no additional cleaning was required.

```python
# Filling missing values if any are present

for col in df.select_dtypes(include='number').columns:
    df[col] = df[col].fillna(df[col].median())

for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].fillna(df[col].mode()[0])
```

Even though my dataset does not contain missing values, I included this step because it is a common part of the data preprocessing process. If any missing values were present, I would replace numerical values with the median because it is less affected by extreme values. For categorical columns, I would use the mode, as it replaces missing values with the most frequently occurring category.

Relationship Between Features
Country identifies where the data belongs.
Year helps compare changes over time.
Energy Type shows the renewable energy source, such as solar, wind, or hydro.
Installed Capacity and Production are closely related because a higher installed capacity usually results in greater energy production.
Investments can influence the growth of renewable energy projects.
GDP may affect a country's ability to invest in renewable energy.
CO₂ Emissions can be compared with renewable energy production to understand environmental impact.
Population influences the overall demand for energy.

These relationships help in understanding how different factors work together in the renewable energy sector.

```python
# Creating a new feature to measure production efficiency

df["Production per MW"] = df["Production (GWh)"] / df["Installed Capacity (MW)"]

df.head()
```

To gain additional insight, I created a new column called Production per MW. This feature represents how much electricity is produced for every megawatt of installed capacity. It helps compare the efficiency of renewable energy production across different records instead of looking only at the total production.
